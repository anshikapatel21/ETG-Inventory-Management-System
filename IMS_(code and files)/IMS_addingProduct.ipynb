Loading JSON file and reading data

```python
import json
fd = open("records.json",'r')
r = fd.read()
fd.close()
r
```

```python
records = json.loads(r)
records
```

Adding New Item

```python
pr_id = int(input("Enter the Product ID : "))
pr_name = input("Enter Product name : ")
price = int(input("Enter price of Product : "))
mfg_date = input("Enter mfg Date of Product : ")
exp_date = input("Enter expiry Date of Product : ")
quant = int(input("Enter quantity : "))
```

```python
records[pr_id] = {"pr_name":pr_name , "price" : price, "mfg_date" : mfg_date, "exp_date": exp_date, "quant": quant}
records
```

Updating Inventory

```python
records[pr_id] = {"pr_name":pr_name , "price" : price, "mfg_date" : mfg_date, "exp_date": exp_date, "quant": quant}

js = json.dumps(records)

fd = open("records.json",'w')
w= fd.write(js)
fd.close() 
```
